# Categorical Data Handling
* The algorithm can only understand numbers hence conversion of categorical data is important 
* Methods to handle categorical data
 <br> 1 using loc function and domain analysis
 <br> 2 One hot encoding
 
 <br> sudo code for loc function:
 <br> df.loc[df[column] == '',column] = 'some numerical value'
 <br> Why particular value taken reason should be reasonable
 

```python
df.loc[df['sex']=='male','sex'] = 0
```

```python
df.loc[df['sex']=='female','sex'] = 1
```

```python
df
```

```python
df.smoker.value_counts()
df.loc[df['smoker']=='yes','smoker'] = 1
df.loc[df['smoker']=='no','smoker'] = 0
```

```python
df.region.value_counts()
```

```python
df.loc[df['region']=='southeast','region'] = 3
df.loc[df['region']=='southwest','region'] = 2
df.loc[df['region']=='northwest','region'] = 1
df.loc[df['region']=='northeast','region'] = 0
```

```python
df
```

# One Hot Encoding

```python
df.smoker.value_counts()
```

```python
df2 = pd.get_dummies(df['smoker'],prefix='smoker',drop_first=True)
df = pd.concat([df,df2],axis=1).drop(['smoker'],axis=1)
```

```python
df2
```

```python
df
```

```python
df3 = pd.get_dummies(df['sex'],prefix='sex',drop_first=True)
df = pd.concat([df,df3],axis=1).drop(['sex'],axis=1)
```
